```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
## to avoid encoding errors, we use unicode_escape
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
## Drop unrealted columns
df.drop(columns=['Status', 'unnamed1'],inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Amount'].mean()
```

```python
## There are 12 null values in Amount, so we will be dropping them
df.dropna(inplace=True)
```

```python
df.shape
```

```python
## change data type 
df['Amount']=df['Amount'].astype(int)
```

```python
df.describe()
```

## Exploratory Data Analysis



#### Gender

```python
plt.figure(figsize=[5,4])
ax = sns.countplot( x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df['Gender'].value_counts()
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=[5,4])
ax=sns.barplot(data=sales_gen, x='Gender', y='Amount')

for bars in ax.containers:
    ax.bar_label(bars)
```

- From above graphs we can see that most of the buyers are Females and even the purchasing power of Females are greater than Men

### Age

```python
df.columns
```

```python
cp=sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in cp.containers:
    cp.bar_label(bars)
```

```python
## Total Amount vs Age Group
sales_age=df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
plt.figure(figsize=[5,4])
sns.barplot(data=sales_age, x='Age Group', y='Amount')
```

- From above graphs we can see that most of the buyers are from age group between 26-35 yrs female

## State

```python
df.columns
```

```python
# total number of orders from top 10 states
import warnings
warnings.filterwarnings('ignore')

sales_state = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

colors=['#FF0000','#0000FF','#00FFFF','#FFFF00','#FF00FF']
plt.figure(figsize=[6,4])
sns.barplot(data=sales_state, x='State', y='Orders',palette=colors)
plt.xlabel('State',weight='bold')
plt.ylabel('Orders',weight='bold')
```

```python
# Total amount/sales from top 10 states

sales_state=df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head()
sns.barplot(data=sales_state, x='State', y='Amount')
```

- This are the top 5 states ranked according to their purchasing power (Amount and Orders)

## Marital Status

```python
plt.figure(figsize=[5,4])
ax= sns.countplot(data=df, x='Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=sales_data, x='Marital_Status', y='Amount', hue='Gender')
```

* From above graph, we can see that most of the buyers are married (women) and they have high purchasing power 

## Occupation

```python
plt.figure(figsize=[20,5])
ax=sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data=df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=[20,5])
sns.barplot(data=sales_data, x='Occupation', y='Amount')
```

From the above graphs, we can see that most of the buyers are working in IT, Healthcare and aviation sector.

## Product Category

```python
plt.figure(figsize=[25,5])

ax=sns.countplot(data=df,x='Product_Category')
plt.xlabel('Product_Category',weight='bold')
plt.ylabel('count',weight='bold')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data=df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=[20,5])
sns.barplot(data=sales_data, x='Product_Category',y='Amount')
```

- From above graphs, we can see that most of the sold products are from Food, Footwear and Electronics Category

```python
sales_data=df.groupby('Product_ID', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=[20,5])
sns.barplot(data=sales_data, x='Product_ID', y='Orders')
```

## Conclusion



Married women age group 25-35 yrs from Uttar Pradesh, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics Industry.
